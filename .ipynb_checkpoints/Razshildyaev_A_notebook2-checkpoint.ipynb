## 1.3.1 Задание
***Задача:***</br> Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и
единицами.
</br> ***Решение:***

```python
mx = np.arange(64)
for x in range(64):
    mx[x] = (x / 8 + x % 8) % 2
mx = mx.reshape(8, 8)
print(mx)
```

## 1.3.2 Задание
***Задача:***</br> Создать 5x5 матрицу со значениями в строках от 0 до 4. Для создания
необходимо использовать функцию arrange
</br> ***Решение:***

```python
mx = np.arange(25)
for x in range(25):
    mx[x] = x % 5
mx = mx.reshape(5, 5)
print(mx)
```

## 1.3.4 Задание
***Задача:***</br> Создать матрицу с 0 внутри, и 1 на границах.
</br> ***Решение:***

```python
w = 6
h = 8
mx = np.arange(w*h)
for x in range(w*h):
    mx[x] = x % w % (w-1) == 0 or x // w % (h-1) == 0
mx = mx.reshape(h, w)
print(mx)
```

## 2.2.1 Пример
***Задача:***</br> Создать Series из списка Python, словаря Python, и массива Numpy
(установить буквенные метки для последнего).
</br> ***Решение:***

```python
import pandas as pd
import numpy as np
lst = [1, 2, 3, 4, 5]
d = {'a':1, 'b':2, 'c':3}
ndarr = np.array([1, 2, 3, 4, 5])

s1 = pd.Series(lst)
s2 = pd.Series(d)
s3 = pd.Series(ndarr, ['a', 'b', 'c', 'd', 'e'])
print(s1)
print(s2)
print(s3)
```

## 2.3.1 Задача
***Задача:***</br> Найдитн евклидово расстояние между двумя Series (точками) a и b, не используя встроенную формулу.
</br> ***Решение:***

```python
def e_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))


a = pd.Series([12, 7, 6])
b = pd.Series([2, 4, 3])
print(e_distance(a, b))
```

## 3.2.1 Пример
***Задача:***</br> Прошкалируйте числовой признак в диапазон между двумя значениями.
</br> ***Решение:***

```python
import numpy as np
from sklearn import preprocessing
feature = np.array([[-500.5], [-100.1], [0], [100.1], [900.9]])
minmax_scale = preprocessing.MinMaxScaler(feature_range = (0, 1))
scaled_feature = minmax_scale.fit_transform(feature)
scaled_feature
```

## 3.2.2 Пример
***Задача:***</br> Преобразуйте признак, чтобы он имел среднее значение 0 и стандартное
отклонение 1
</br> ***Решение:***

```python
x = np.array([[-1000.1], [-200.2], [500.5], [600.6], [9000.9]])
scaler = preprocessing.StandardScaler()
standardized = scaler.fit_transform(x)
display(standardized)
print('Среднее:', round(standardized.mean()))
print('Стандартное отклонение:', standardized.std())
```

## 3.2.3 Пример
***Задача:***</br> Дан фрейм данных
dfTest = pd.DataFrame({'A':[14.00,90.20,90.95,96.27,91.21],
'B':[103.02,107.26,110.35,114.23,114.68],
'C':['big','small','big','small','small']})
Необходимо масштабировать его числовые столбцы
</br> ***Решение:***

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dfTest = pd.DataFrame({'A':[14.00,90.20,90.95,96.27,91.21], 
                       'B':[103.02,107.26,110.35,114.23,114.68], 
                       'C':['big','small','big','small','small']})
dfTest[['A', 'B']] = scaler.fit_transform(dfTest[['A', 'B']])
dfTest
```

## 3.3.2 Задача
***Задача:***</br> Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.
</br> ***Решение:***

```python
url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
dataframe = pd.read_csv(url)
minmaxscaler = preprocessing.MinMaxScaler(feature_range = (0, 1))
dataframe[['sepal_length_cm']] = minmaxscaler.fit_transform(dataframe[['sepal_length_cm']])
zscaler = preprocessing.StandardScaler()
dataframe[['sepal_width_cm']] = zscaler.fit_transform(dataframe[['sepal_width_cm']])
dataframe.head(5)
```
